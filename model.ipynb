```python
from concurrent.futures import ThreadPoolExecutor
# Function to get artist id from track id
def get_artist_id(track_id):
    track_info = sp.track(track_id)
    if len(track_info['artists']) > 0:
        artist_id = track_info['artists'][0]['id']
        return artist_id
    else:
        return None

# Remove NaN values from track_id column
merged_df = merged_df.dropna(subset=['track_id'])

# Create a ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=20) as executor:
    # Apply the get_artist_id function to retrieve artist_id for each track_id
    merged_df['artist_id'] = list(executor.map(get_artist_id, merged_df['track_id']))
```

```python
def get_artist_ids(df):
    df = df.dropna(subset=['track_id'])

    chunks = np.array_split(df['track_id'], len(df) // 50 + 1)

    artist_ids = [None] * len(df)

    for chunk in chunks:
        tracks_info = sp.tracks(chunk.tolist())
        
        for track_info in tracks_info['tracks']:
            index = df.index[df['track_id'] == track_info['id']][0]
            artist_ids[index] = track_info['artists'][0]['id'] if len(track_info['artists']) > 0 else None

    return artist_ids

merged_df['artist_id'] = get_artist_ids(merged_df)
```

```python
def merge_datasets(track_data, album_data, artist_data):
    # Renaming the columns
    artist_data.rename(columns={'id': 'artists_id', 'name': 'artists_name'}, inplace=True)
    album_data.rename(columns={'id': 'album_id', 'release_date': 'release_date'}, inplace=True)

    # Merging track_data and album_data on 'album_id'
    merged_df = pd.merge(track_data, album_data[['album_id', 'release_date']], on='album_id')

    # Merging merged_df and artist_data on 'artists_id'
    merged_df = pd.merge(merged_df, artist_data[['artists_id', 'artists_name', 'genres']], on='artists_id')

    # Extracting the release year from release_date
    merged_df['release_year'] = pd.to_datetime(merged_df['release_date']).dt.year

    # Selecting the columns you want
    merged_df = merged_df[['artists_id', 'acousticness', 'danceability', 'disc_number',
       'duration_ms', 'energy', 'href', 'id', 'instrumentalness', 'key',
       'liveness', 'loudness', 'lyrics', 'mode', 'name', 'playlist',
       'popularity', 'preview_url', 'speechiness', 'tempo', 'time_signature',
       'track_href', 'uri', 'valence', 'release_date', 'artists_name',
       'genres', 'release_year']]
    
    return merged_df
final_df = merge_datasets(track_data, albums_data, artist_data)
```

## Get Filtered_Data

```python
## join artist genre information and album release date with track dataset
# drop irrelevant columns
# get only tracks after 1990
def join_genre_and_date(artist_df, album_df, track_df):
    album = album_df.rename(columns={'id':"album_id"}).set_index('album_id')
    artist = artist_df.rename(columns={'id':"artists_id",'name':"artists_name"}).set_index('artists_id')
    track = track_df.set_index('album_id').join(album['release_date'], on='album_id' )
    track.artists_id = track.artists_id.apply(lambda x: x[2:-2])
    track = track.set_index('artists_id').join(artist[['artists_name','genres']], on='artists_id' )
    track.reset_index(drop=False, inplace=True)
    track['release_year'] = pd.to_datetime(track.release_date).dt.year
    track.drop(columns = ['Unnamed: 0','country','track_name_prev','track_number','type'], inplace = True)
    
    return track[track.release_year >= 1990]
```

```python
def get_filtered_track_df(df, genres_to_include):
    df['genres'] = df.genres.apply(lambda x: [i[1:-1] for i in str(x)[1:-1].split(", ")])
    df_exploded = df.explode("genres")[df.explode("genres")["genres"].isin(genres_to_include)]
    df_exploded.loc[df_exploded["genres"]=="korean pop", "genres"] = "k-pop"
    df_exploded_indices = list(df_exploded.index.unique())
    df = df[df.index.isin(df_exploded_indices)]
    df = df.reset_index(drop=True)
    return df
```

```python
genres_to_include = genres = ['dance pop', 'electronic', 'electropop', 'hip hop', 'jazz', 'k-pop', 'latin', 'pop', 'pop rap', 'r&b', 'rock']
track_with_year_and_genre = join_genre_and_date(artist_data, albums_data, track_data)
filtered_track_df = get_filtered_track_df(track_with_year_and_genre, genres_to_include)
```
